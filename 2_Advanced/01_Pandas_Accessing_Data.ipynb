```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()
```

```python
#Clean Data
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

# iloc

Using df.iloc[]: Select rows and columns based on position
Syntax: df.iloc[rows, columns]. Alternatively you can use ":" to slice rows or columns

Examples of using iloc:
- Get the first row (index 0)
- Get the job_skills element of the 3rd (index at 2)
- Get the job_skills (index 15) and job_type_skills (index 16) for the third (index 2) and fourth (index 3) rows
- Get the first 12 rows of the DataFrame
- Get the first five columns of the DataFrame and all of the rows

```python
df.iloc[0]
```

```python
df.iloc[15,16]
```

```python
df.iloc[[2,3]]
```

```python
df.iloc[[2,3], [15,16]]
```

```python
#First 10 rows of the DataFrame
#df.iloc[:9]
df.iloc[0:10]
```

```python
#Get the first 5 columns of the DataFrame and all of the rows
df.iloc[:, :5]
```

# loc
df.loc: Select rows and columns by position or label
Syntax: df.loc[rows, "columns"]. Alternatively you can use ":" to slice rows or columns
Similar to df.iloc[] except we can use labels instead

Examples
- Get the first row (index 0)
- Get the first 10 rows of job_skills and job_type_skills
- Get the first 5 columns and rows 10-20
- Get the first 12 rows of the DataFrame
- Get the first 5 columns of the DataFrame and all the rows

```python
#Get the first row
df.loc[0]
```

```python
#Get the first 10 rows of job_skills and job_type_skills
df.loc[:9,["job_skills", "job_type_skills"]]
#Alternative - df.loc[:9][["job_skills", "job_type_skills"]]
```

```python
#Get the first 5 columns and rows 10-20
df.loc[10:20, 'job_title_short':'job_work_from_home']
```

```python
df.loc[:, "salary_rate":"salary_hour_avg"].dropna(subset='salary_rate')
```
