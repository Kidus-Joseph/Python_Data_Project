# Copy
copy() will make a copy of the object's indices and data

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_original
```

```python
df_altered = df_original

df_altered
```

```python
df_altered.loc[:5, "salary_year_avg"]
```

```python
median_salary = df_altered['salary_year_avg'].median()
```

```python
median_salary
```

```python
df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(value=median_salary)
```

```python
df_altered.loc[:5, "salary_year_avg"]
```

```python
df_original.loc[:5, 'salary_year_avg']
```

# Why did df_altered and df_original have the same result?

Because we used the "=" operator, we copied the id of the df_original variable to the df_altered variable
Therefore, the variables are referencing the same dataframe

```python
id(df_original)
```

```python
id(df_altered)
```

```python
dataset = load_dataset('lukebarousse/data_jobs')
df_original = dataset['train'].to_pandas()
```

```python
df_altered = df_original.copy()
```

```python
print("ID of OG dataset : ", id(df_original))
print("ID of Altered dataset: ", id(df_altered))
print("Are they the same dataframes? : ", id(df_original) == id(df_altered))
```

```python
median_salary = df_altered['salary_year_avg'].median()

df_altered['salary_year_avg'] = df_altered['salary_year_avg'].fillna(value=median_salary)

df_altered.loc[:5, 'salary_year_avg']
```
