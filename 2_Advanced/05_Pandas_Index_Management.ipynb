# Three Methods for Indexes

- reset_index()
- set_index()
- sort_index()

### reset_index()

- Resets the index, or a level of it

```python
df_usa = df[df['job_country'] == 'United States']
df_usa
```

```python
df_usa.reset_index(inplace=True)
df_usa
```

### set_index

- Set the DataFrame index using existing columns
- Set the DataFrame index (row labels) using one or more existing columns or arrays (of the correct length). The index can replace the existing or expand on it

```python
df_usa.set_index('job_index', inplace=True)
df_usa
```

### sort_index()

- Sort object by labels (along an axis)
- Returns a new DataFrame sorted by label if inplace argument is False, otherwise updates the original DataFrame and returns None
- This function is best used when grouping or pivoting data

```python
median_pivot = df_usa.pivot_table(values='salary_year_avg', index='job_title_short', aggfunc=['median', 'min', 'max'])
median_pivot
```

```python
median_pivot.sort_values(by=('median', 'salary_year_avg'), ascending=False)
```

```python
median_pivot.sort_index(inplace=True)
median_pivot
```

```python
median_pivot.sort_values(by='job_title_short', ascending=False)
median_pivot
```

### When would you use sort_index() function?

- You could perhaps use this function when the index column doesn't have a title
