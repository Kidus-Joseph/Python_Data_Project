# Conditional Statements

If
1.   Purpose - First condition check
2.   Condition Evaluated - First condition to be evaluated
3.   Action if True - Executes block if condition is true

Elif (you can have multiple elif statements)
1.   Purpose - Subsequent condition check
2.   Condition Evaluated - Next condition, if "if" is false
3.   Action if True - Executes block if this condition is true

Else
1.   Purpose - Fallback if no conditions are met
2.   Condition Evaluated - None
3.   Action if True - Executes block if all above conditions are false





Start conditional statement with keyword, in this case "If"

If

```python
if False:
  print("What's up Data Nerds")
```

```python
#Skill required for job posting
job_skill = "SQL"

#Applicant Skill
applicant_skill = "SQL"

if job_skill == applicant_skill:
  print("Skills match!")
```

If - Elif

```python
job_skill = "SQL"
applicant_skill = "Python"
years_experience = 3

if job_skill == applicant_skill:
  print("Skills match!")
elif years_experience >= 5:
  print("Enough Experience; No SQL")
```

If - Elif - Else

```python
job_skill = "SQL"
applicant_skill = "Python"
years_experience = 3

if job_skill == applicant_skill:
  print("Skills match!")
elif years_experience >= 5:
  print("Enough Experience; No SQL")
else:
  print("No Skill or Experience")
```

Example with multiple Elif statements

```python
job_skill = "SQL"
applicant_skill = "Python"
years_experience = 3

if job_skill == applicant_skill:
  print("Skills match!")
elif years_experience >= 5:
  print("Enough Experience; No SQL")
elif applicant_skill == "Python":
  print("This applicant should know SQL")
else:
  print("No Skill or Experience")
```

- Sometimes you may see the keyword, Pass
- Pass is used when you don't know what value you want to put there for the time being but you need something to be within or underneath that keyword if or elif
- Ellipses (...) can also be used in place of the word "Pass"

```python
job_skill = "SQL"
applicant_skill = "Python"
years_experience = 3

if job_skill == applicant_skill:
  print("Skills match!")
elif years_experience >= 5:
  print("Enough Experience; No SQL")
elif applicant_skill == "Python":
  pass
else:
  print("No Skill or Experience")
```
