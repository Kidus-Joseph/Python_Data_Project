# Modules

Modules

- Most commonly used to define functions and call them
- To call a function, use the following syntax
  - module_name.function_name()
- Example: math.sqrt

```python
import my_module

my_module.skill_list
```

```python
my_module.skill("Python")
```

Real world example
- Import isn't the only way to call a function, especially if you are only using one or two functions
- You can use the From and Import key word
  - This means you will not have to use the dot notation
- You can also use the * after import to import all the functions that belong to the modules

```python
from job_analyzer import *

calculate_salary(100000)
calculate_bonus(110000, 100000)
```

Because it is a function belonging to a module, you can now use the help function to learn about the what the function does
- You can also write descriptions explaining what it does
   - You do this by writing """ on two lines and write your description in between those lines with """ on top and """ on the bottom

```python
help(calculate_salary)
```

```python
help(calculate_bonus)
```

# Python Standard Library

Statistics Module

```python
salary_list = [98000, 101000, 102000, 99000, 97000]
```

```python
import statistics

statistics.mean(salary_list)
```

```python
from statistics import mean, mode, median
mean(salary_list)
mode(salary_list)
median(salary_list)
```
