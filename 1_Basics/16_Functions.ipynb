# Function

There are several types of functions
- Built in Functions
  - Already built and provided by Python
  - Example: print()
- User-defined functions
  - Functions that you build yourself from scratch
  - Example: def my_function(): pass
- Lambda functions
  - Example: lambda x: x + 1
- Standard Library functions
  - Example: math.sqrt()
- Third-Party Library Functions
  - Example: numpy.array()

# User Defined Functions

```python
base_salary = 100000
bonus_rate = 0.1

# Calculate total_salary of base_salary * (1 + bonus_rate)
# Use the pass keyword at the end of your function (indented inside) to make sure it is running correctly
# When you want to result, use the return keyword to state what it is that you want
# The variables you use can be stated prior the function being defined or inside the function being called
# The variables don't have to be stated above the function. Important thing to note is that the function must be defined before it is called
# You can also provide arguments by including it in the function definition line at the top of the function

def calculate_salary(base_salary, bonus_rate):
  #base_salary = 100000
  #bonus_rate = 0.1
  total_salary = base_salary * (1 + bonus_rate)

  return total_salary
```

```python
# When calling the function, you must provide arguments
calculate_salary(base_salary, bonus_rate)
```

```python
# The arguments you provide do not have to be variables. They can be numbers, strings, etc. depending on the function being called
calculate_salary(500000, 0.25)
```

In the below example, you can define the value of variable by assigning a value right next to the argument itself (look at bonus_rate)

```python
def calculate_salary1(base_salary, bonus_rate=0.1):

  total_salary = base_salary * (1 + bonus_rate)

  return total_salary
```

If you decide to provide an argument, that argument provided will override the argument you assigned a value to. If not, it will use the assigned value above

```python
salary_1 = 100000
rate_1 = 0.5

calculate_salary1(salary_1, rate_1)
```
