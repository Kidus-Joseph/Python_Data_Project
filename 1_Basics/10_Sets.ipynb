# Sets

Sets are denoted with curly brackets {} and you don't need to use colons

Sets are unique in the fact that if you converted a list with duplicate values, sets will drop the duplicate value

- Usage: For unique items
- Syntax: {item1, item2,..}
- Type of Data: Set
- Order: Unordered
- Indexing: No
- Duplicate values: Not allowed
- Mutability: Mutable

You can put any datatype into a set

```python
job_skills = {'tableau', 'sql', 'python', 'statistics'}
```

```python
job_skills
```

You can't index a set meaning they are not subscriptable

```python
job_skills[1]
```

```python
help(set)
```

```python
job_skills.add('looker')
```

```python
job_skills
```

Testing to see if you can add a value that is already in the set

```python
job_skills.add('sql')
```

```python
job_skills
```

The pop method for sets does not take arguments

It is going to provide an arbitary value and remove it

```python
job_skills.pop("tableau")
```

```python
job_skills.pop()
```

```python
job_skills
```

If you wanted to remove a specific value from the set, you have to use the remove method

```python
job_skills.remove('tableau')
```

```python
job_skills
```

Using sets, you can extract out unique values and do so in an efficient manner

```python
skill_list = ['python', 'sql', 'statistics', 'tableau', 'python', 'sql', 'statistics', 'tableau']
```

```python
list(set(skill_list))
```
