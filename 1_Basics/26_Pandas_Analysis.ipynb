# Methods for Statistical Understading of Columns

- df.count(): Count of values
- df.sum(): Sum of values
- df.cumsum(): Cummulative sum of values
- df.idxmin()/df.idxmax(): Indexes of min/max value
- df.median(): Median of values
- df.mode(): Mode of values
- df.column_name.value_counts(): Unique counts

```python
df.count()
```

# Groupby function

Syntax: dataframe_name.groupby("columns to group by")["columns to aggregate"].your_agg_method()

```python
df.groupby('job_title_short')['salary_year_avg'].median()
```

The below doesn't work because you have to put the square brackets around the column names. Anytime you want to include multiple columns, they have to be in a list

```python
df.groupby('job_title_short', 'job_country')['salary_year_avg'].median()
```

Providing multiple columns you want to group by

```python
df.groupby(['job_title_short', 'job_country'])['salary_year_avg'].median()
```

Providing mulitple columns you want to aggregate by

```python
df.groupby('job_title_short')[['salary_year_avg', 'salary_hour_avg']].median()
```

Performing multiple aggregation methods
- This is done by using agg function and writing down the aggregate functions in a list

```python
df.groupby('job_title_short')['salary_year_avg'].agg(['min', 'max', 'median'])
```
