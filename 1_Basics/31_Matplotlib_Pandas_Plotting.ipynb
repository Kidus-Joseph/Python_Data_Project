```python
import pandas as pd
import matplotlib.pyplot as plt
from datasets import load_dataset
```

```python
#Load Data
dataset = load_dataset("lukebarousse/data_jobs")
df = dataset["train"].to_pandas()

#Cleanup
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

# Bar Chart - Counts of Job Postings

```python
job_counts = df.job_title_short.value_counts()

job_counts
```

```python
plt.bar(job_counts.index, job_counts)
plt.title("Postings by Job Title")
plt.ylabel("Count of Job Postings")
plt.xticks(rotation=45, ha='right')
plt.show()
```

# Plotting Through Pandas
- Two options
- pd.Series.plot
- pd.DataFrame.plot

Plotting through a Series

```python
job_counts.plot(kind='bar')
```

```python
job_counts.plot(kind='bar')
plt.title('Job Postings by Job Title')
plt.xlabel('') #how to remove title
plt.ylabel('Count of Job Postings')
plt.xticks(rotation=45, ha='right')
plt.show()
```

Plotting through a Dataframe
