# Tuples

Tuples are denoted with parentheses () and it can contain any data type

Unlike lists, if you wanted to modify the tuple, you wouldn't be able to . Not even add or remove items

- Usage: For fixed data (ex: unique set of items that you don't want altered)
- Syntax: (item1, item2,..) or item,
- Type of Data: Sequence
- Order: Ordered
- Indexing: Yes
- Duplicate values: Allowed
- Mutability: Immutable

Tuples are more applicable to software engineering

```python
lukes_skill = ('python', 'sql', 'statistics', 'tableau')
lukes_skill
```

Tuples are indexable and can be sliced

Index: tuple_name[0]
Slicing: tuple_name[start:end"step]

```python
lukes_skill[0]
```

```python
lukes_skill[0:2]
```

```python
lukes_skill.append('r')
```

```python
help(tuple)
```

```python
lukes_skill = ('python', 'sql', 'statistics', 'tableau')

new_skills = ('excel', 'power bi')

lukes_skill + new_skills
```

Although tuples are immutable, you can technically still add to a tuple

The only difference is that whenver you do modify an existing tuple, you will be creating an entierly new object

You can use the id function, id(), to look properly identify an object and get its proper unique value

```python
lukes_skill = ('python', 'sql', 'statistics', 'tableau')

print(id(lukes_skill))

new_skills = ('excel', 'power bi')

lukes_skill += new_skills

lukes_skill
```

```python
id(lukes_skill)
```

Unlike tuples, lists will have the same value showing that one list can replace the other, whereas with tuples, a new one is created

```python
lukes_skill = ['python', 'sql', 'statistics', 'tableau']

print(id(lukes_skill))

new_skills = ['excel', 'power bi']

lukes_skill += new_skills

lukes_skill
id(lukes_skill)
```
