# Membership Operators

Membership operators check to see if whether an object is inside of another object

- In
  - x in y
- Not In
  - x not in y

Membership operators are case-sensitive

```python
"Data" in "Data Nerd"
```

```python
"data" in "Data Nerd"
```

```python
job_skills = ['python', 'r', 'sql']
```

```python
'python' in job_skills
```

```python
'python' not in job_skills
```

# Identity Operators

Identity operators are used to see if an item is correlated to another item

- IS
  - x is y
- IS NOT
  - x is not y

The IS operator is commonly confused with the equal operator (==) to see if two values are equal to one another which is not its purpose. The purpose of the IS operator is to check whether or not these two variables are using the same memory location for the value that they are holding

```python
salary_kelly = 100_000
salary_luke = 100_000

salary_kelly == salary_luke
```

```python
salary_kelly is salary_luke
```

```python
id(salary_kelly)
```

```python
id(salary_luke)
```

```python
id(salary_kelly) == id(salary_luke)
```
