# Loops

Loops


*   for Loops
    - Use case: Iterates over a sequence (like a list, tuple, dictionary, string, or range)
    - Control: It runs for a specific number of iterations, defined by the sequence or range
    - Best for: When you know in advance how many times you need to loop
      - Examples: Iterating over a sequnce, fixed number of iterations, Processing each character in a string, Iterating over keys and values in a dictionary, List comprehension, Nested Iteration, When the Range is Finite
*   while Loops
    - While loops will check for conditions on whether it is true or false and while it is maintaining true the condition underneath will continue to run. Use While loops when the number of iterations is not predetermined, or you need to keep looping until a specific condition is met.
    - Use case: Repeats as long as condition is true
    - Control: It keeps running until the specified condition becomes false
    - Best for: When the number of iterations isn't known in advance, but depends on a condition
      - Examples: Repeating Until a Condition is Met, Handling User Input or Actions, When the Loop Condition May Change During Execution, Infinite Loops (with Break Condition), Waiting for an Event or Condition, Uncertain Number of Iterations, Looping Based on a Complex Condition, Resource Monitoring or Polling
*   Key Differences
    - Iteration control
      - for Loop: Pre-determined number of iterations based on a sequence or range
      - while Loop: Number of iterations depends on the condition and could be indefinite if the condition doesn't change
    - Typical Use:
      - for: Used when iterating over a collection (list, string, range)
      - while: Used when waiting for a condition to be met
    - Risk of Indefinite Loop:
      - for Loops generally do not result in infinite loops
      - while Loops can lead to infinite loops if the condition never becomes false

```python
years_experience = {
    'Luke': 3,
    'Kelly': 6,
    'Ken': 4,
    'Ann': 3,
    'Mary': 7
}
```

```python
years = 0
index = 0

years_list = list(years_experience.items())

while years < 5:
  key, years = years_list[index]
  if years >= 5:
    print(key, "has", years, "years of experience")
  index += 1
```
