You can use a comparison operator to filter the dataset
- You can also filter by rows by applying a condition inside brackets [].
- With this syntax: df[df['column_name'] > value].
- The > can be any conditional operator like: >, <, ==, etc.

```python
df.job_title_short == "Data Analyst"
```

The code inside the brackets is the condition

```python
df[df.job_title_short == "Data Analyst"]
```

The below is ambiguous because we need to put the conditions in parentheses

```python
df[df.job_title_short == "Data Analyst" and df.salary_hour_avg > 100000]
```

The below is stil ambiguous because we are use the comparison word, not bitwise operator symbol

```python
df[(df.job_title_short == "Data Analyst") and (df.salary_hour_avg > 100000)]
```

```python
df[(df.job_title_short == "Data Analyst") & (df.salary_year_avg > 100000)]
```

```python
df[(df.job_title_short == "Data Analyst") | (df.salary_year_avg > 100000)]
```

Find Not NA Values
- pd.notna() - displays if the values are not NA (missing values).
- It returns a Boolean same-sized object indicating if the values are not NA.
- Non-missing values are True
- Missing values are False
- Good for data preprocessing to remove or fill missing values, or to make decisions based on the presence of data.

```python
df.salary_year_avg.notna()
```

```python
df[(df.job_title_short == "Data Analyst") & (df.salary_year_avg.notna())]
```
