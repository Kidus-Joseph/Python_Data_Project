# Variables: The Reference to Objects

Variable: A defined name that references an object

```python
job_title = "Data Analyst"
job_location = "United States"
job_salary = 90000
```

You can find that a variable directly points to an object by using something like the id function. This will generate an identification number to see if it is correlated to that object

```python
help(id)
```

```python
id(job_title)
```

```python
id(job_location)
```

Just because two different variables hold the same value doesn't mean they will be the same. They will still point to different objects

```python
job_1 = "Data Analyst"
job_2 = "Data Analyst"
```

```python
id(job_1)
```

```python
id(job_2)
```

However, if you were to make a variable and have another variable as it's value then they will both be the same because they are now pointing to the same object

```python
job_3 = job_1
```

```python
id(job_3), id(job_1)
```

You can also use variables on functions as well

```python
print("Hello")
```

```python
my_print_func = print
```

```python
my_print_func("Hello")
```

# Functions: The Manipulators of Objects

Function: A reusable piece of code that performs a specific task; manipulators of different objects

To run a function, you need a parenthese

```python
print
```

```python
print()
```

```python
print("Hello")
```

The print function is a built in function; however if you wanted to use the function to print out a specific statement over and over again. You can build your own function

```python
def greet():
  return "What's up?"
greet()
```

```python
print(f"Job: {job_title}\nLocation: {job_location}\nSalary: ${job_salary: ,.0f}")
```

```python
def display_info(title, location, salary):
  return print(f"JOB: {title}\nLocation: {location}\nSalary: ${salary:,.0f}")
```

```python
display_info(job_title, job_location, job_salary)
```

# Classes: The Template of Objects

Class: A template for creating objects (records)

```python
type(90000)
```

```python
help(int)
```

Int is a class and 90000 is an instance of that class

```python
class JobPost:
  def __init__(self, title, location, salary) :
    self.title = title
    self.location = location
    self.salary = salary
```

```python
JobPost(job_title, job_location, job_salary)
```

The above is now an instance of the class JobPost

# Attributes: The variables of an Object

Attributes: A field in a record, defined by its class

Using the above class JobPost, we have actually already defined attributes - title, location, and salary

```python
job_1 = JobPost(job_title, job_location, job_salary)
```

You can access attributes of an instance of a class by using dot notation. You don't need to use the parentheses

```python
job_1.title
```

```python
job_1.location
```

# Methods: The Functions of an Object

Method: A function defined inside a class that operates on its objects
