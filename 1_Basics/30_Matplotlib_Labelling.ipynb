```python
#Loading Dataset
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset["train"].to_pandas()
```

```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
```

```python
#Cleaning
df["job_posted_date"] = pd.to_datetime(df["job_posted_date"])
```

# Bar Chart - Counts of Job Countings

```python
job_counts = df.job_title_short.value_counts()
job_counts
```

```python
plt.bar(job_counts.index, job_counts)
plt.show()
```

# Adding a title to the chart

```python
plt.bar(job_counts.index, job_counts)
plt.title("Postings by Job Title")
plt.show()
```

# Adding a Y label

```python
plt.bar(job_counts.index, job_counts)
plt.title("Postings by Job Title")
plt.ylabel("Count of Job Postings")
plt.show()
```

# Adding/Editing a X Label

```python
plt.bar(job_counts.index, job_counts)
plt.title("Postings by Job Title")
plt.ylabel("Count of Job Postings")
#plt.xlabel("...") Creating an x axis
plt.xticks(rotation = 45)
plt.show()
```

```python
plt.bar(job_counts.index, job_counts)
plt.title("Postings by Job Title")
plt.ylabel("Count of Job Postings")
plt.xticks(rotation = 45, ha = 'right') #rotation will rotate x axis and ha, aka horizontal alignment, will adjust how
plt.show()
```
