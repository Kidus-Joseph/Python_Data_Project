# Lambda

AKA Anonymous functions. Very simiilar in how they operate to user defined function and built in function. More of an advanced functionality.

Lambda functions Syntax
- lambda x: x + 1
- lambda arguments: expression



You can set the lambda function equal to an equation

```python
mul_two = lambda x: x * 2
```

You can call the function using the variable you've assigned it to and provide values as arguments like the below

```python
mul_two(3)
```

The point of a lambda function is to be able to write everything out in one line

```python
(lambda x: x * 2)(3)
```

You can pass multiple arguments through a lambda function

```python
(lambda x, y : x * y)(3, 4)
```

Previous Example with User-Defined functions

```python
salary_list = [100000, 200000, 150000, 120000, 80000, 75000]

#Calculate total_salary of base_salary * (1 + bonus_rate)
def calculate_salary(base_salary, bonus_rate = 0.1):

  return base_salary * (1 + bonus_rate)

total_salary_list = [calculate_salary(salary) for salary in salary_list]

total_salary_list
```

Same example but with Lambda

```python
salary_list = [100000, 200000, 150000, 120000, 80000, 75000]

#Calculate total_salary of base_salary * (1 + bonus_rate)
def calculate_salary(base_salary, bonus_rate = 0.1):

  return base_salary * (1 + bonus_rate)

total_salary_list = [(lambda x: x * 1.1)(salary) for salary in salary_list]

total_salary_list
```

# Data Filtering Example

Find jobs that have Python and that are remote

```python
jobs_data = [
    {"job_title": 'Data Scientist', 'job_skills': ['Python', 'Machine Learning'], 'remote': True},
    {'job_title': 'Data Analyst', 'job_skills': ['Excel', 'SQL'], 'remote': False},
    {'job_title': 'Machine Learning Engineer', 'job_skills': ['Python', 'TensorFlow', 'Keras'], 'remote': True},
    {'job_title': 'Software Developer', 'job_skills': ['Java', 'C++'], 'remote': True},
    {'job_title': 'Data Scientist', 'job_skills': ['R', 'Statistics'], 'remote': False}
]
```

```python
help(filter)
```

```python
filter(lambda job: job['remote'], jobs_data)
```

```python
list(filter(lambda job: job['remote'], jobs_data))
```

```python
list(filter(lambda job: job['remote'] and "Python" in job['job_skills'], jobs_data))
```
