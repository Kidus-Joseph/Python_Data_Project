# List Comprehension

- Syntax: ["expression" for "item" in "iterable"]
- The first part is the variable you want to put in
-

```python
numbers = [x for x in range(100)]
numbers
```

Here is how to perform an operation on the list comprehension
- You will edit the expression part of the syntax

```python
numbers = [x*2 for x in range(100)]
numbers
```

```python
numbers = [float(x) for x in range(100)]
numbers
```

You can use any item that is iterable such as list, sets, tuples, dictionaries, range functions, strings, etc.

```python
numbers = [float(x) for x in range(100)]
numbers
```

```python
numbers = [x for x in 'PYTHON']
numbers
```

# Types of comprehension

You are not only limited to list comprehension
- You can do set, tuple, and dictionary comprehension
- All you have to do is change the brackets

```python
numbers = {x for x in 'PYTHON'}
numbers
```

```python
job_list = [
    "Business Analyst",
    "Data Analyst",
    "Financial Analyst",
    "Data Scientist",
    "Market Research Analyst",
    "Operations Analyst",
    "Machine Learning Engineer",
    "Risk Analyst",
    "Systems Analyst",
    "Research Analyst",
    "Investment Analyst",
    "Data Engineer",
    "Credit Analyst",
    "Management Analyst",
    "Product Analyst",
    "Supply Chain Analyst",
    "Pricing Analyst",
    "Healthcare Analyst",
    "Budget Analyst",
    "Quantitative Analyst",
    "Policy Analyst",
    "Fraud Analyst",
    "Compliance Analyst",
    "AI Specialist",
    "Data Architect",
    "Business Intelligence Analyst",
    "Statistical Analyst",
    "Data Warehouse Engineer",
    "ETL Developer",
    "Big Data Engineer",
    "Database Administrator",
    "Data Pipeline Engineer",
    "Cloud Data Engineer"
]
```

```python
analyst_list = []

for job in job_list:
  if "Data Analyst" in job:
    analyst_list.append(job)
```

```python
analyst_list = [job for job in job_list]
analyst_list
```

```python
len(analyst_list)
```

```python
analyst_list = [job for job in job_list]
```

```python
analyst_list
```

```python
len(analyst_list)
```

```python
len(job_list)
```

```python
print("Original list is:", len(job_list), "jobs")
print("New list is:", len(analyst_list), "jobs")
```

If you wanted to add an if statement to the list comprehension, you just have to add it at the end, behind the iterable

```python
analyst_list = [job for job in job_list if "Data Analyst" in job]
```

```python
print("Original list is:", len(job_list), "jobs")
print("New list is:", len(analyst_list), "jobs")
```
